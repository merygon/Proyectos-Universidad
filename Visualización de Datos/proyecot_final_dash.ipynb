```python
df1 = pd.read_csv("parte_1.csv")
df2 = pd.read_csv("parte_2.csv")
df = pd.concat([df1, df2])
```

```python
tabs_styles = {
    'height': '44px'
}
tab_style = {
    'borderBottom': '1px solid #d6d6d6',
    'padding': '6px',
    'fontWeight': 'bold'
}

tab_selected_style = {
    'borderTop': '1px solid #d6d6d6',
    'borderBottom': '1px solid #d6d6d6',
    'backgroundColor': '#119DFF',
    'color': 'white',
    'padding': '6px'
}
```

## Estudio de los datos del df

```python
# Datos necesarios para las grficas del apartado a de la pestaa 1:
#df["date"] = pd.to_datetime(df["date"])
num_tiendas = df["store_nbr"].value_counts()
tiendas = df["store_nbr"].unique()
num_productos = df["family"].value_counts()
productos = df["family"].unique()
num_estados = df["state"].value_counts()
estados = df["state"].unique()
num_meses = df["month"].value_counts()
meses = df["month"].unique()

# Datos necesarios para las grficas del apartado b de la pestaa 1:
# Dado que nos piden un analisis por trmino medio, para no fastidiar los datos originales, creamos un dataframe nuevo con los mismos datos y una columna aadida, que representa las ventas medias
df_copy = df
mean_sales = []
for sale in df["sales"]:
    mean = np.mean(sale)
    mean_sales.append(mean)
df_copy["mean_sales"] = mean_sales
df_pmv = df_copy.sort_values("mean_sales", ascending = False)

# Datos necesarios para las grficas del apartado c de la pestaa 1:

df_dia_ventas = df[["day_of_week","sales"]].groupby("day_of_week").mean().sort_values(by="sales",ascending=False) # dataframe que contiene las ventas por da de la semana
df_semanas_ventas = df[["week","sales"]].groupby("week").mean().sort_values(by="sales",ascending=False) # dataframe que contiene las ventas por semana del ao
df_meses_ventas = df[["month","sales"]].groupby("month").mean().sort_values(by="sales",ascending=False) # dataframe que contiene las ventas por mes del ao
```

```python
# Pgina 1

tab1_content = html.Div([
    dcc.Tabs(id="tabs-example-graph-1", value='tab-1', children=[
        # Apartado a
        html.H3('Conteo general'),
        html.Div([
            html.Div([dcc.Graph(figure=px.bar(x = tiendas, y = num_tiendas, title = "Nmero Total de tiendas"))]),
            html.Div([dcc.Graph(figure=px.bar(x = productos, y = num_productos, title = "Nmero Total de Productos que se venden"))]),
            html.Div([dcc.Graph(figure=px.bar(x = estados, y = num_estados, title = "Estados en los que est la empresa"))]),
            html.Div([dcc.Graph(figure=px.bar(x = meses, y = num_meses, title = "Meses disponibles"))]),
        ]),
        # Apartado b
        html.H3("Anlisis en trminos medios"),
        html.Div([
            html.Div([dcc.Graph(figure=px.bar(x = df_pmv["family"], y = df_pmv["mean_sales"], title = "Ranking (top 10) de los produtos ms vendidos"))]),
            html.Div([dcc.Graph(figure=px.histogram(x = df_pmv["store_nbr"], y = df_pmv["mean_sales"], title = "Distribucin de ventas por tiendas"))]),
            html.Div([dcc.Graph(figure=px.bar(x = df_pmv["store_nbr"], y = df_pmv["onpromotion"], title = "Ranking (top 10) de tiendas con ventas en produtos con promocin"))])
        ]),
        # Apartado c
        html.H3("Anlisis de estacionalidad en trminos medios"),
        html.Div([
            html.Div([dcc.Graph(figure=px.bar(x = df_meses_ventas.index, y = df_meses_ventas["sales"], title = "Da de la semana con ms ventas"))]),
            html.Div([dcc.Graph(figure=px.bar(x = df_semanas_ventas.index, y = df_semanas_ventas["sales"], title = "Semana con ms ventas"))]),
            html.Div([dcc.Graph(figure=px.bar(x = df_dia_ventas.index, y = df_dia_ventas["sales"], title = "Mes con ms ventas"))])
            ])
    ]
    )
])

# Paso 2: Creacin del layout del Dashboard
app = dash.Dash(__name__) # Llamamos a la app de dash para poder crear una app con un link

app.title = "Estudio de datos" # Ttulo general del Dashboard

# Creacin del layout: 
app.layout = html.Div([
    html.H1('Estudio de datos'),
    dcc.Tabs(id="tabs-example-graph-1", value='tab1_content', children=[
        # Apartado 1
        dcc.Tab(label='Conteo General', value='tab1_content', style=tab_style, selected_style=tab_selected_style)
        ]),
    html.Div(id='tabs-content-example-graph')
])

@app.callback(
    Output('tabs-content-example-graph', 'children'),
    Input('tabs-example-graph-1', 'value'))
def render_content(tab):
    return tab1_content

# Paso 4: Ejecutamos la aplicacin
app.run_server(debug=True,jupyter_mode="external")
```
